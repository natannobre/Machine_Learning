# Regressão Linear

```python
from sklearn import linear_model
from sklearn import metrics
from matplotlib import pyplot as plt
import numpy as np
```

```python
X = np.array([[1], [2], [4], [3], [5]])
y = np.array([1, 3, 3, 2, 5])
```

```python
X
```

```python
plt.grid()
plt.scatter(X, y);
```

### Instanciando e usando o modelo implementado pelo scikit learn

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X, y)
```

### Modelo ajustado tem importantes informações: coeficientes e ponto de interceptação com o eixo y.

```python
model.coef_
```

```python
model.intercept_
```

```python
f'{model.intercept_} + {model.coef_[0]} * x'
```

```python
y_pred = model.predict(X)
y_pred
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred)
plt.plot(X, y_pred, 'o');
```

### MSE - Mean Squared Error - Erro Médio Quadrático

```python
np.sum((y - y_pred) ** 2) / 5
```

```python
metrics.mean_squared_error(y, y_pred)
```

### RMSE - Root Mean Squared Error - Raiz do Erro Médio Quadrático

```python
(np.sum((y - y_pred) ** 2) / 5) ** 0.5
```

```python
metrics.mean_squared_error(y, y_pred, squared=False)
```

### O desvio padrão de y nos dá uma ideia de que o RMSE deve ser bem melhor (e menor) do que ele.

```python
np.std(y)
```

### Se usássemos a média de y como predição (y_pred), teríamos...

```python
np.mean(y)
```

```python
y.shape
```

```python
y_pred = np.full((5,), 2.8)
y_pred
```

```python
(np.sum((y - y_pred) ** 2) / 5) ** 0.5
```

```python
metrics.mean_squared_error(y, y_pred, squared=False)
```

### MAE - Mean Absolute Error - Erro Médio Absoluto

```python
np.sum(np.abs(y - y_pred)) / 5
```

```python
metrics.mean_absolute_error(y, y_pred)
```
