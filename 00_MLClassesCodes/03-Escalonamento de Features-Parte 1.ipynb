```python
import numpy as np
import matplotlib.pyplot as plt
```

## Distribuição Normal ou Gaussiana

```python
s = np.random.normal(size=10000, loc=0, scale=10)
s
```

```python
s.min()
```

```python
plt.hist(s, bins=50);
```

```python
s = np.random.normal(size=10000, loc=0, scale=1)
plt.hist(s, bins=50);
```

```python
np.random.seed(0)
x = np.random.rand(20)
x = (x * 100).round(2)
x = np.resize(x, (20, 1))
```

## Distribuição não normal

```python
x
```

```python
plt.hist(x, bins=20);
```

```python
np.mean(x[:, 0])
```

```python
np.std(x[:, 0])
```

## Padronização ou Standardização de Features


```python
(x[:, 0] - np.mean(x[:, 0])) / np.std(x[:, 0])
```
