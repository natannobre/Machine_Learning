```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('https://github.com/regispires/aulas-machine-learning-2019/blob/master/datasets/planilha1.xlsx?raw=true')
df
```

```python
map_tam = { 'size':  { 'S': 0, 'M': 1, 'L': 2, 'XL': 3 } }
```

```python
df_replace = df.replace(map_tam)
df_replace
```

```python
one_hot = pd.get_dummies(df_replace['color'], drop_first=True)
one_hot
```

```python
df_replace.drop(['color'], axis=1, inplace=True)
```
