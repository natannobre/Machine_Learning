### Assinalar valores aos faltantes

```python
df_ = df.dropna(how='all')
df_
```

```python
# aqui o eixo 0 é o eixo das colunas
prep = SimpleImputer()
prep.fit(df_)
dados = prep.transform(df_)
dados
```
