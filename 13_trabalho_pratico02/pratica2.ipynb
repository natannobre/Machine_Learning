```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.tree import DecisionTreeRegressor,DecisionTreeClassifier
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

# Generate and plot a synthetic imbalanced classification dataset
from numpy import where
from collections import Counter
import imblearn
from imblearn.over_sampling import SMOTE
```

# 3. Pré-processamento dos dados

```python
bankrupt.groupby('Bankrupt?').count()
```

```python
y = bankrupt['Bankrupt?'].values
print(y.shape)
X = bankrupt.copy()
del X['Bankrupt?']
X = X.values
X.shape
```

```python
# summarize class distribution
counter = Counter(y)
print(counter)
```

## 3.2. SMOTE - Synthetic Minority Oversampling Technique

```python
# transform the dataset
oversample = SMOTE(sampling_strategy='minority', random_state=42, k_neighbors=3) # Default: k_neighbors=5
X_smote, y_smote = oversample.fit_resample(X, y)
# summarize the new class distribution
counter = Counter(y_smote)
print(counter)
```

# 4. Modelos preditivos de classificação

```python
X_train, X_test, y_train, y_test = train_test_split(X_smote, y_smote, train_size=0.8, random_state=42, stratify=y_smote)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
knn = KNeighborsClassifier()
svm = SVC()
dt = tree.DecisionTreeClassifier(random_state=42)
rf = RandomForestClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)
mlpc = MLPClassifier(random_state=0)
```

### Grid Search : KNeighborsClassifier

```python
params_knn = {'n_neighbors':np.arange(1, 15)}
grid_knn = GridSearchCV(knn, param_grid=params_knn, cv=5, scoring='f1_weighted')
grid_knn.fit(X_train, y_train)
grid_knn_f1 = grid_knn.best_score_
print(grid_knn.best_params_)
print(grid_knn_f1)
```

### Grid Search : SVM

```python
params_svm = {'kernel' : ['rbf', 'sigmoid'], 'cache_size' : [2, 5, 10, 15, 25, 50, 100]}
grid_svm = GridSearchCV(svm, param_grid=params_svm, cv=5, scoring='f1_weighted')
grid_svm.fit(X_train, y_train)
grid_svm_f1 = grid_svm.best_score_
print(grid_svm.best_params_)
print(grid_svm_f1)
```

### Grid Search : DecisionTreeClassifier

```python
params_dt = {'max_depth' : np.arange(1, 30)}
grid_dt = GridSearchCV(dt, param_grid=params_dt, cv=5, scoring='f1_weighted')
grid_dt.fit(X_train, y_train)
grid_dt_f1 = grid_dt.best_score_
print(grid_dt.best_params_)
print(grid_dt_f1)
```

### Grid Search : RandomForestClassifier

```python
params_rf = {'n_estimators' : [100, 200, 300, 400, 500, 600, 700]}
grid_rf = GridSearchCV(rf, param_grid=params_rf, cv=5, scoring='f1_weighted')
grid_rf.fit(X_train, y_train)
grid_rf_f1 = grid_rf.best_score_
print(grid_rf.best_params_)
print(grid_rf_f1)
```

### Grid Search : GradientBoostingClassifier

```python
params_gb = {'n_estimators' : [200, 300, 400]}
grid_gb = GridSearchCV(gb, param_grid=params_gb, cv=5, scoring='f1_weighted')
grid_gb.fit(X_train, y_train)
grid_gb_f1 = grid_gb.best_score_
print(grid_gb.best_params_)
print(grid_gb_f1)
```

### Grid Search : Multi-layer Perceptron

```python
params_mlpc = {'max_iter' : [200, 300, 400, 500]}
grid_mlpc = GridSearchCV(mlpc, param_grid=params_mlpc, cv=5, scoring='f1_weighted')
grid_mlpc.fit(X_train, y_train)
grid_mlpc_f1 = grid_mlpc.best_score_
print(grid_mlpc.best_params_)
print(grid_mlpc_f1)
```

# Apresentação dos Resultados

```python
def hist_graph(y,title):
    labels = ['KNeighbors', 'SVM', 'Decision Tree', 'Random Forest', 'GradientBoosting', 'Multi Layer Perceptron']
    fig = px.bar(x=labels, y=y, color=labels, title=title, text=y,
        labels=dict(x="Algorithms", y="F1 Score", color="Algorithm")
    )
    fig.update_traces(texttemplate='%{text:.2%}', textposition='outside')
    fig.show()
```

```python
hist_graph(y=[grid_knn_f1, grid_svm_f1, grid_dt_f1, grid_rf_f1, grid_gb_f1, grid_mlpc_f1],title='F1 Score com GridSearch')
```

### Sem Grid Search

```python
train_list = []

no_grid_knn = KNeighborsClassifier().fit(X_train,y_train)
train_list.append(no_grid_knn)

no_grid_svm = SVC().fit(X_train,y_train)
train_list.append(no_grid_svm)

no_grid_dt = DecisionTreeClassifier(random_state=42).fit(X_train,y_train)
train_list.append(no_grid_dt)

no_grid_rf = RandomForestClassifier(random_state=42).fit(X_train,y_train)
train_list.append(no_grid_rf)

no_grid_gb = GradientBoostingClassifier(random_state=42).fit(X_train,y_train)
train_list.append(no_grid_gb)

no_grid_pc = MLPClassifier(random_state=42).fit(X_train,y_train)
train_list.append(no_grid_pc)
```

```python
f1_list = []
for i in train_list:
    y_pred_test = i.predict(X_test)
    f1_list.append(metrics.f1_score(y_test, y_pred_test))

hist_graph(y=f1_list,title='F1 Score sem GridSearch')
```
